```python
books_cleaned=books.drop_duplicates()
books_cleaned
```

```python
user_books_cleaned=books_cleaned.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
user_books_cleaned
```

```python
user_books_cleaned.index=books_cleaned['User.ID'].unique()
user_books_cleaned
```

```python
user_books_cleaned.fillna(0, inplace=True)
user_books_cleaned
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1-pairwise_distances(user_books_cleaned.values,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index=books['User.ID'].unique()
user_sim_df.columns=books['User.ID'].unique()
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
